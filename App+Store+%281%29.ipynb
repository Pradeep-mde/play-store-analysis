```python
# Q4

duplicates = df.duplicated()
print(f'Number of duplicate rows = %d' % (duplicates.sum()))
```

```python
def find_column_duplicates(df):
    duplicates = {}
    for column in df.columns:
        column_duplicates = df[column].duplicated(keep=False)
        if column_duplicates.any():
            duplicates[column] = df[column][column_duplicates].tolist()
    return duplicates

column_duplicates = find_column_duplicates(df)
print("Column-wise duplicates:")
print(column_duplicates)
```

```python
df = df.drop_duplicates()
df
```

```python
# Q4

unique_categories = df['Category'].unique()
unique_categories = [category for category in unique_categories if category != '1.9']
df = df[df['Category'] != '1.9']
unique_categories
```

```python
df.isnull().sum()
```

```python
# Q5
missing_values = df['Rating'].isnull().sum()
print("Number of missing values in 'Rating' column:", missing_values)
```

```python
df = df.dropna(subset=['Rating'])


df['Rating_category'] = df['Rating'].apply(lambda x: 'High' if x > 3.5 else 'Low')


print(df.head())
```

```python
category_distribution = df['Rating_category'].value_counts()
print(category_distribution)
```

```python
# Q8

df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
sns.boxplot(x=df['Reviews'])
plt.show()

```

```python
df['Reviews_log'] = np.log1p(df['Reviews'])

# Plotting the transformed data
sns.boxplot(x=df['Reviews_log'])
plt.show()
```

```python
df.head()
```

```python
# Q9
df['Size'] = df['Size'].replace({'M': 'e+6', 'K': 'e+3', 'Varies with device': np.nan})


df['Size'] = pd.to_numeric(df['Size'], errors='coerce')


df = df.dropna(subset=['Size'])

print(df.head())  
```

```python
# Q10
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')

df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')


print(df.head())
```

```python
# Q11
df['Price'] = df['Price'].str.replace('$', '')

df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

print(df.head())
```

```python
# Q12
redundant_columns = ['Rating', 'App', 'Genres', 'Last Updated', 'Current Ver', 'Android Ver']
df = df.drop(columns=redundant_columns)

print(df.head())
```

```python
# Q13

from sklearn.preprocessing import LabelEncoder


label_encoder = LabelEncoder()


for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = label_encoder.fit_transform(df[column])


print(df.head())
```

```python
df
```

```python
# Q14

target = df['Rating_category']
independent_features = df.drop(columns=['Rating_category'])

print("Target (Rating_category):")
print(target.head())

print("\nIndependent Features:")
print(independent_features.head())
```

```python
# Q15

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(independent_features, target, test_size=0.2, random_state=42)


print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Q16
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Display the scaled data
print("Scaled X_train:")
print(X_train_scaled)

print("\nScaled X_test:")
print(X_test_scaled)
```

```python
print(f'Shape of the dataframe = {df.shape}')
columns = df.shape[1]
rows = df.shape[0]
print(f'Total number of columns in the dataset = {columns}')
print(f'Total number of rows in the dataset = {rows}')
```

```python
df.dtypes
```

```python
df['Year'] = df['year']


year_distribution = df['Year'].value_counts(normalize=True) * 100

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(year_distribution, labels=year_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of Data by Year')
plt.axis('equal') 
plt.show()
```

## Inference



```python
plt.figure(figsize=(10, 6))
sns.histplot(df['priceperlb'], kde=True)
plt.title('Distribution of Price per lb')
plt.xlabel('Price per lb')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the scatterplot between 'numcol' and 'prodvalue'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='numcol', y='prodvalue', data=df)
plt.title('Relationship between Number of Colonies and Production Value')
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value')
plt.show()
```

### Inference


* Positive Relationship: There appears to be a positive relationship between the number of colonies and production value. As the number of colonies increases, there is a tendency for the production value to increase as well.

* Presence of Outliers: There are a few data points that deviate from the general trend, particularly at higher production values. These outliers may represent cases where factors other than colony numbers significantly influence production value

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Relationship between Year and Production Value')
plt.xlabel('Year')
plt.ylabel('Production Value')
plt.show()
```

# inference

Variation Over Time: The boxplot shows the distribution of production values ('prodvalue') for each year. Look for any patterns or trends in the distribution across different years. Are there any noticeable changes or fluctuations in production values over time

Median and Spread: For each year, examine the median (middle line in the box), as well as the spread of production values (length of the box and whiskers). This provides insights into the central tendency and variability of production values for each year.

Outliers: Identify any outliers in the boxplot. Outliers are data points that deviate significantly from the rest of the data and may indicate unusual or extreme production values for a particular year.



```python
columns_of_interest = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']

# Filter the DataFrame to include only the selected columns
df_selected = df[columns_of_interest]

# Plot pairplot
sns.pairplot(df_selected, hue='year', diag_kind='kde')
plt.suptitle('Pairplot of Variables with Different Years')
plt.show()
```

# Inference

Distribution of Variables: The diagonal plots (kde plots) show the distribution of each variable. Examine how the distribution of each variable changes over different years. Are there any noticeable shifts or changes in distribution patterns

Relationship Between Variables: The scatterplots show the relationship between pairs of variables. Look for any patterns or trends in how the variables are related to each other across different years. Are there any consistent relationships between certain pairs of variables, or do these relationships vary over time

Yearly Variation: Pay attention to how the relationships between variables change from one year to another. Are there any consistent changes in these relationships across different years.For example, do certain variables tend to be positively or negatively correlated in some years but not in others

```python
columns_of_interest = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']

# Calculate correlation matrix
correlation_matrix = df[columns_of_interest].corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```

# Inference


Strength of Correlation: Examine the color intensity of the cells in the heatmap. Darker shades indicate stronger correlations (either positive or negative), while lighter shades indicate weaker correlations or no correlation.

Direction of Correlation: Look at the signs of the correlation coefficients (positive or negative). A positive correlation coefficient indicates a positive linear relationship between two variables, while a negative correlation coefficient indicates a negative linear relationship.

Interpretation of Correlation Coefficients: Analyze the numerical values in the cells. Correlation coefficients close to 1 or -1 indicate strong correlations, while coefficients close to 0 suggest weak or no correlation.
